```python
import pandas as pd
notas = pd.read_csv("/content/ratings.csv")
notas.columns = ["usuarioId","filmeID","Nota","momento"]
notas["Nota"].value_counts()

frequencia = notas["Nota"].value_counts().max()
nota = notas["Nota"].value_counts().idxmax()

import seaborn as sns
sns.histplot(notas["Nota"])
notas.query("Nota > 3")
```
